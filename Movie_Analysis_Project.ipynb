```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Jupyter 
%matplotlib inline
plt.rcParams['figure.figsize']= (10,6)

df=pd.read_csv("C:\\Users\\omkar\\Downloads\\Movies_Analysis.csv")
df.head()
```

# 1. Basic Exploration

```python
# shape, columns, dtypes, basic info
print("Shape:", df.shape)
print("\nColumns:\n", df.columns.tolist())
df.info()
```

```python
# unique counts and earliest/latest year 
print("\nUnique movies:", df['title'].nunique() if 'title' in df else "No title column")
for col in ['genre','director','lead_actor','country','release_year']:
    if col in df.columns:
        print(f"{col} unique:", df[col].nunique())
if 'release_year' in df.columns:
    print("Earliest:", df['release_year'].min(), "Latest:", df['release_year'].max())
```

# 1.1 Missing values And summary statistics

```python
# Missing Values
missing = df.isnull().sum().sort_values(ascending=False)
missing[missing>0]
```

```python
# Shape (rows Ã— columns)
print("Shape:", df.shape)

# Column names
print("\nColumns in dataset:")
print(df.columns.tolist())

# Data types
print("\nData Types:")
print(df.dtypes)

# Basic info
print("\nInfo:")
df.info()
```

```python
# summary stats for numeric cols
df.describe().T
```

# 2.Univariate analysis

```python
# release year histogram
if 'release_year' in df.columns:
    df['release_year'].dropna().astype(int).hist(bins=30)
    plt.title('Distribution of Release Years')
    plt.xlabel('Year')
    plt.ylabel('Count')
    plt.show()
```
